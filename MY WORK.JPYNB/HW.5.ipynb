```python
products = ["Laptop", "Mobaile","Table", "Headphone"]
sales = [54000,20000,5000,30000]
plt.bar(products, sales, color = 'Green')
plt.title("Sales by porduct name")
plt.xlabel("products")
plt.ylabel("sales")

plt.grid(axis='y')
plt.show()
```

```python
product  = ['Laptop','Mobaile', 'Table','Headphone']
sales = [45000,20000,25000,1000]
plt.barh(product , sales, color = 'yellow')
plt.title('Sales by porduct name')
plt.xlabel("product")
plt.ylabel("sales")
plt.grid(axis='y')
plt.show()
```
