```python
# ine chart

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# step 1:
x =  [1,2,3]
y = [10,20,30]

# step 2: create plot
plt.plot(x,y)
plt.plot("Simple Line Graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
# step 3:show output
plt.show()
```

```python
# Date
days =  [1,2,3,4,5]
sales = [100,150,120,200,180]

# plot
plt.plot(days , sales , marker='*' ,color='orange', linestyle ='dashed')

plt.title("Sales Trend Over 5 Days")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

## Sales by product category

```python
# data
products = ["Laptop", "Mobaile","Tablet","Headphones"]
sales = [40000,70000,30000,20000]
plt.bar(products,sales, color = 'pink')
plt.title("Sales by product category")
plt.xlabel("Products")

plt.ylabel("Sales")
plt.grid(axis='y')
plt.show()
```

## pie chart

```python
# "Monthly Expences distribution
cat = ['rent','food','travel','savings']
expenses = [15000,8000,4000,3000]

plt.pie(
    expenses,
    labels = cat,
    autopct = '%1.1f%%',
    startangle = 90
)
plt.title("Monthly Expences distribution")
plt.show()
    
```

```python
explode = [0,0,0.1,0]
colors = ["blue", "green", "yellow","purple"]

plt.pie(
    expenses,
    labels = cat,
    explode=explode,
    autopct = '%1.1f%%',
    colors = colors,
    startangle = 90
)
plt.show()
```

```python
# data
days = [1,2,3,4,5]
sales = [100,150,200,180,250]
profit = [20,30,50,40,60]
customers = [10,20,30,25,35]

fig,ax = plt.subplots(2,2)

#line plot
ax[0,0].plot(days,sales,color = 'orange',marker ='o')
ax[0,0].set_title("Sales Trend")

# barchart
ax[0,1].barh(days,profit,color = "purple")
ax[0,1].set_title("Profit comparison")

# scatterplot
ax[1,0].scatter(customers,sales,s=100,color = "red")
ax[1,0].set_title("Coustomers vs sales")

# historgram
ax[1,1].hist(sales,bins=4 , edgecolor="black",color="pink")
ax[1,1].set_title("Sales Distribution")

plt.tight_layout()
plt.show()
```
