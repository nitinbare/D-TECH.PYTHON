```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## load dataset

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
print(df.head())
```

```python
# scater plot
sns.set_style('darkgrid')
sns.scatterplot(x ='total_bill',y='tip',data=df,hue = 'sex')
plt.show()
```

```python
sns.lineplot(x = 'total_bill',y='tip',data = df)
plt.show()
```

```python

sns.set_style('darkgrid')
sns.scatterplot(x ='total_bill',y='tip',data=df,hue = 'smoker')
plt.show()
```

## bar plot

```python
# bar plot show average (mean) value of categories
```

```python
sns.barplot(x ='day',y='total_bill', data= df,hue = 'sex')
plt.show()
```

```python
# a count plot show frequency (count of records)
```

```python
sns.countplot(x = 'day' ,data = df ,hue = 'sex')
plt.show()
```

```python
## histogram
```

```python
# it show how data is distributed
```

```python
sns.histplot(df['total_bill'],kde = True)
plt.show()
```

```python
#kernel density estimate
```

```python
##KDE plot
```

```python
#a KDE(kernel Density Estimate show a smooth curve of distribution
```

```python
sns.kdeplot(df['total_bill'],fill = True)
plt.show()
```

```python
# box plot
sns.boxplot(x='day',y='total_bill',data = df,hue = 'sex')
plt.show()
```

```python
## violin plot
# it is combination of box plpot and KDE plot
# it show distribution + density
```

```python
sns.violinplot(x ='day',y ='total_bill',data=df)
plt.show()
```

```python
## pair plot
# a pair plot show relationship between all numerical columns
```

```python
sns.pairplot(df,hue = 'sex')
plt.show()
```

```python
## heatmap
# a heatmap show correlation between variable using colors
```

```python
corr = df.corr(numeric_only = True)
sns.heatmap(corr, cmap = 'coolwarm', annot = True)
plt.show()
```
