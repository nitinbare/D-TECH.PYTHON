```python
# outlier detection
data = np.array([10,12,11,13,100])
Q1 = np.percentile(data,25)
Q3 = np.percentile(data,75)
print("Q1:", Q1)
print("Q2:" , Q3)

IQR = Q3 - Q1
print("IQR:" , IQR)

lower = Q1 - 1.5*IQR
print("Lower:", lower)
upper = Q3 + 1.5*IQR
print("Upper:", upper)

outlier = data[(data<lower) | (data>upper)]
print("outliers:" , outlier)
```
