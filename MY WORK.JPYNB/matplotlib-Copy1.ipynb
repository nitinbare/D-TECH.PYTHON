```python
# ine chart

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# step 1:
x =  [1,2,3]
y = [10,20,30]

# step 2: create plot
plt.plot(x,y)
plt.plot("Simple Line Graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
# step 3:show output
plt.show()
```

```python
# Date
days =  [1,2,3,4,5]
sales = [100,150,120,200,180]

# plot
plt.plot(days , sales , marker='*' ,color='orange', linestyle ='dashed')

plt.title("Sales Trend Over 5 Days")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

## Sales by product category

```python
# data
products = ["Laptop", "Mobaile","Tablet","Headphones"]
sales = [40000,70000,30000,20000]
plt.bar(products,sales, color = 'pink')
plt.title("Sales by product category")
plt.xlabel("Products")

plt.ylabel("Sales")
plt.grid(axis='y')
plt.show()
```

```python
# data
products = ["Laptop", "Mobaile","Tablet","Headphones"]
sales = [40000,70000,30000,20000]
plt.barh(products,sales, color = 'pink')
plt.title("Sales by product category")
plt.xlabel("Products")

plt.ylabel("Sales")
plt.grid(axis='y')
plt.show()
```
