## indexing in 1D array

```python
a = np.array([10,20,30,40])
# first element
print("First lement:" ,a[0])
# last element           
print("Last Elemet:" ,a[-1])

```

```python
# slicing in  1D Array
a = np.array([10,20,30,40,50])
print("extract 20 to 40:" , a[1:4])
print("Extract form starting to 30:" , a[:3])
print("Extract from 40 to end: " ,a[3:])
print("extract by step 2:" ,a[::2])
print("Reverse array:",a[::-1])
```

```python
# indexing  in 2D
a = np.array([[1,2,3], [4,5,6]])
print("Extract 2 element:" ,a[0,1])
print("Extract 6:",a  [1,2])

# slicing in 2D
print("second column:",a[:,1])
print("first row:" ,a[0,:])
print("extract 2,3 and 5,6:",a[:,1:])      
      
```

```python
# fancy indexing
a = np.array([10,20,30,40,50])
print
```

```python
# fancy indexing
a = np.array([10,20,30,40,50,])
print("Extract only 10 and 30:",a[[0,2]])
      
```

```python
# copy vs view
# copy
a = np.array([1,2,3])
b = a.copy()
b[0] = 100
print("original array:" , a)
print("copied array:",b)
print("original array:", a)
```

```python
# view 
import numpy as np
a = np.array([10,20,30])
print("original array:" , a)
b = a.view()
b[0]= 100
print("copied array:", b)
print("original array:", a)
```

```python
# iteration:

# mean accessing elements of an array one by one
```

```python
# basic iteration (1D array)
a = np.array([10,20,30])
for i in a:
    print(i*3)
```

```python
# iteraation 2D array
a = np.array([[1,2],
              [3,4]])
for row in a:
    print(row)
```

```python
# element wise iteration (nested loop)
for row in a:
    for element in row:
        print(element)
```

```python
# np.nditer()- used for efficient iteration in multi demension array

a = np.array([[1,2],
              [3,4]])
for x in np.nditer(a):
    print(x)
```

```python
# iteration with index
a = np.array([[1,2],
              [3,4]])
for index,value in np.ndenumerate(a):
    print(index,value)
```
