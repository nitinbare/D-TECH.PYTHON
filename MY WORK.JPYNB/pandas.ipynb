```python
## pivote table

data = {
    "Name":["Arjun", "Rahul", "Sneha","Amit","Pooja"],
    "Dept":["IT","HR","IT","HR","IT"],
    "Gender":["M","M","F","M","F"],
    "Salary":[50000,400000,600000,3400000,230000]
}
df = pd.DataFrame(data)
print(df)
```

```python
pd.pivot_table(df,values = "Salary", index = "Dept")
```

```python
pd.pivot_table(df,values = "Salary" , index = "Dept",columns = "Gender")
```

```python
pd.pivot_table(df,values = "Salary" , index= "Dept", aggfunc = ["mean","sum"])
```

```python
pd.pivot_table(df,values = "Salary" , index="Dept", columns="Gender",fill_value =0)
```

## MERGING AND JOINING

```python
df1 = pd.DataFrame({
    "ID":[1,2,3],
    "Name":["Arjun","Rahul","Sneha"]
})
df2 = pd.DataFrame({
    "ID":[1,2,3],
    "Salary":[20000,20000,30000]
})
df1
df2
```

```python
# 1. inner join - only matching records
pd.merge(df1,df2,on= "ID")
```

```python
# 2. left join - all rows from left table
pd.merge(df1,df2,on="ID",how="left")
```

```python
# right join - all row from right table
pd.merge(df1,df2,on="ID" , how = "right")
```

```python
# outer join - all rows from from both tables
pd.merge(df1,df2,on="ID" ,how= "outer")
```

```python
# join
df1 = pd.DataFrame({
    "Name":["Arjun","Rahul","Sneha"]
}, index = [1,2,3])

df2 = pd.DataFrame({
    "Salary":[50000,40000,60000]
},index = [1,2,4])
df1
df2
```

```python
# left join
df1.join(df2,how = "left")
```

```python
# right join
df1.join(df2,how = "right")
```

```python
# inner join
df1.join(df2,how = "inner")
```

```python
# outer join
df1.join(df2,how = "outer")
```

```python
data = {
    "Name":["Arjun","Rahul","Sneha","Amit","Pooja"],
    "Dept":["IT","HR","IT", "HR","IT"],
    "Salary":[50000,30000,40000,50000,20000]
}
df = pd.DataFrame(data)
print(df)
```

```python
df["Salary"] =df["Salary"].apply(lambda x:x+50000)
df
            
```

```python
df["Bonus"]= df.apply(lambda x:x["Salary"]*0.1,axis=1)
df
```

```python
def Category(Salary):
    if Salary > 50000:
       return "High"
    else:
       return "Law"
df['Category']=df['Salary'].apply(Category)
df
```
